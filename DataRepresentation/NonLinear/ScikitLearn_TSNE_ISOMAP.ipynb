```python
from sklearn.datasets import load_digits
from sklearn.manifold import TSNE, Isomap
from matplotlib import pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
sns.set(rc={'figure.figsize': (11.7, 8.27)})
palette = sns.color_palette("bright", 10)
```

```python
X, y = load_digits(return_X_y=True)
```

```python
isomap = Isomap()
X_embedded_tsne = TSNE(n_components=2, perplexity=30.0).fit(X)
X_embedded=X_embedded_tsne.embedding_
X_embedded_isomap = isomap.fit_transform(X)
print(f'tsne kl_divergence : {X_embedded_tsne.kl_divergence_}')
```

```python

sns.scatterplot(X_embedded[:, 0], X_embedded[:, 1], hue=y, legend='full', palette=palette).set(title='T-SNE')
plt.show()

sns.scatterplot(X_embedded_isomap[:, 0], X_embedded_isomap[:, 1], hue=y, legend='full', palette=palette).set(title='ISOMAP')
plt.show()
```
